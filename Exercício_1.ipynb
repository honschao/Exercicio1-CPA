```python
import os
import requests
from bs4 import BeautifulSoup
import re
from pathlib import Path
import json
import time
import random
```

## Tarefa 1 ##

```python
def link_collector(verbetes_links_, verbetes_links_visitados, pasta_html, limite, verbetes_links_visitados_titles):
    
    while len(verbetes_links_visitados) < limite:
        
        for link in list(verbetes_links_):

        #   Sem isso não funciona, por algum motivo que não sei 
            if len(verbetes_links_visitados) >= limite:
                
                break  

            if link not in verbetes_links_visitados:

                response = requests.get(link, headers={"User-Agent": "Mozilla/5.0"})
#                 time.sleep(random.uniform(0.5, 1.5))
                verbetes_links_visitados.add(link)

                if response.status_code == 200:
                    # Parsear o HTML com BeautifulSoup
                    soup = BeautifulSoup(response.text, "html.parser")

                    # Pega o título da página e formata como nome de arquivo
                    titulo = soup.title.string if soup.title else "pagina_desconhecida"
                    titulo = re.sub(r'[\\/*?:"<>|]', "", titulo)  # Remove caracteres inválidos
                    titulo = titulo.replace(' – Wikipédia, a enciclopédia livre', '')
                    verbetes_links_visitados_titles.append(titulo)
                    
                    # Cria o nome do arquivo (adicionei um "hash" ao nome de cada arquivo para evitar duplicatas
                    nome_arquivo = os.path.join(pasta_html, f"{len(verbetes_links_visitados)}_{titulo}_verbete.html")
                    
                    with open(nome_arquivo, "w", encoding="utf-8") as arquivo:
                        arquivo.write(response.text)
                    
                    # Encontrar todos os links
                    links_da_pagina = soup.find_all("a")

                    # Para cada link da pagina selecionada (2)filtra para only verbetes e (3)se o link não foi visitado adiciona-o em uma lista com links a serem visitados
                    for novo_link in links_da_pagina:
                        href = novo_link.get("href")
                        # (2) 
                        if href and href.startswith("/wiki/") and ":" not in href:  # Evita páginas de categorias e discussões
                            full_url = "https://pt.wikipedia.org" + href
                        # (3)
                            if full_url not in verbetes_links_visitados:
                                verbetes_links_.append(full_url)  # Adiciona novos links para visitar
    #                                 print(f"{novo_link.get_text()}: {full_url}")  # Exibir na tela

                else:
                    print(f"Erro ao acessar {link}: {response.status_code}")

verbetes_links_ = ["https://pt.wikipedia.org"]  # Começa pela página inicial
verbetes_links_visitados = set() 
pasta_html = 'Verbetes_htmls'
verbetes_links_visitados_titles = []

# Variável que controla a quantidade de páginas visitadas
limite = 5000

link_collector(verbetes_links_, verbetes_links_visitados, pasta_html, limite, verbetes_links_visitados_titles)
```

## Tarefa 2 ##

```python
def InfoboxVerify (pasta_html_):
    
    # Para cada arquivo html em determinada pasta verifica se existe infobox
    for arquivo_html in pasta_html_.glob("*.html"):
        with open(arquivo_html, "r", encoding="utf-8") as f:
            soup = BeautifulSoup(f, "html.parser")
            infobox = soup.select_one("table.infobox")
            
            # Se existir "passe" para a função que extrai e salva os dados
            if infobox:
                
                InfoboxData(infobox)            
            
            else:
            
                pass
      
def InfoboxData (infobox):
    
    dados_infobox = {}

    # Procura o titulo pela tag: <th colspan = 2> para depois nomear o arquivo json
    th_titulo = infobox.find("th", colspan = "2")
    titulo = th_titulo.get_text(strip=True) if th_titulo else "sem_titulo"
    titulo = re.sub(r'[\\/*?:"<>|]', "", titulo)  
    titulo = titulo.replace(" – Wikipédia, a enciclopédia livre", "")
    titulo = titulo.replace("/", "-")  
    
#       Evita travar o script por conta de algum titulo muito loko
    if len(titulo) > 100:
            
        titulo_json_ = os.path.join('Infoboxes_json', f"titulo_desconhecido.json")

    else:

        titulo_json_ = os.path.join('Infoboxes_json', f"{titulo}.json")

    # Percorre toda linha com a tag tr dentro da tag infobox
    for linha in infobox.find_all("tr"):

            # Acha a chave pela tag: <td scope = row>, se tiver chave o valor vai ser o elemento na próxima tag <td>
            chave = linha.find("td", attrs={"scope": "row"})
            
            if chave: 
                
                valor = chave.find_next_sibling('td')
            
            else:
                
                continue
            
            if chave and valor:
                
                chave_texto = chave.get_text(separator=" ", strip=True)
                
                # Verifica se é uma lista
                itens = valor.find_all("li")
                
                if itens:
                    
                    lista_valores = [item.get_text(separator=" ", strip=True) for item in itens]
                    dados_infobox[chave_texto] = lista_valores
                
                else:
                    
                    dados_infobox[chave_texto] = valor.text.strip()
    
    with open(titulo_json_, "w", encoding="utf-8") as f:
        json.dump(dados_infobox, f, ensure_ascii=False, indent=4)

pasta_html_ = 'Verbetes_htmls'
InfoboxVerify(Path(pasta_html_))
```
